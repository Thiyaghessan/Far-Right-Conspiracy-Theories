## Research Question: How effective are account bans at stopping the spread of conspiracy theories on social media?

### Proposed Method: Interactive Game

### Step 1: Load Python Scripts

```python
import conspiracy_model as m
import networkx as nx
```

### Step 2: Initialize Network

 - You can change the parameters for the network in the conspiracy_model script

```python
model = m.VirusOnNetwork()
nx.draw(model.G, with_labels=True)
```

### Step 3: Play Game

**Rules**

 - First you specify the number of turns you would like the game to run
 - At the start of each round, some nodes will be radicalised (state.RADICALISED, color=Red) based on parameters in conspiracy model
 - The rest of the nodes will be innocent citizens (state.CITIZEN, color=lightblue)
 - Each round you have a choice to ban a particular node from the social network
  - Doing so runs the risk of radicalising that node's neighbours, since the act of banning an account can induce outrage with probability 0.3, however banned nodes can no longer infect neighbours
  - if you do not choose to ban a node, that node will try and infect its neighbours each turn with probability 0.3
 - At the same time, radicalised nodes have the chance to deradicalise
  - if no new developments happen in the news cycle (probability=0.4), individuals can get bored and revert to state.CITIZEN with probability 0.2 (so overall chance == 0.4 * 0.2 = 0.08)
  - Alternatively radicalised individuals can turn into counter radicals with probaility 0.1 (state.IMMUNE, color=pink)
 - The aim is for the player to decide if they want to ban nodes or let the conspiracy theory spread and let individuals decide if they want to believe these theories or not

```python
model.run_model(10)
```
